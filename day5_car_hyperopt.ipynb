```
df = pd.read_hdf('car.h5')
df.shape 
```

## Future Engineering 

```
SUFFIX_CAT = '__cat'
for feat in df.columns:
  if isinstance(df[feat][0], list) : continue

  factorized_values = df[feat].factorize()[0]
  if SUFFIX_CAT in feat:
    df[feat] = factorized_values
  else:
    df[feat + SUFFIX_CAT] = factorized_values
```

```
df['param_rok-produkcji'] = df['param_rok-produkcji'].map(lambda x: -1 if str(x) == 'None' else int(x) )
df['param_moc'] = df['param_moc'].map(lambda x: -1 if str(x) == 'None' else int(str(x).split(' ')[0] ))
df['param_pojemność-skokowa'] = df['param_pojemność-skokowa'].map(lambda x: -1 if str(x) == 'None' else int( str(x).split ('cm3')[0].replace(' ','')))
```

### Function

```
def run_model(model,feats):

  X = df[feats].values
  y = df['price_value'].values

  scores = cross_val_score(model,X,y, cv=3, scoring='neg_mean_absolute_error')
  return np.mean(scores), np.std(scores)
```

```
feats =  ['param_napęd__cat',   
          'param_stan__cat',
          'param_faktura-vat__cat',   
          'param_rok-produkcji',
          'param_skrzynia-biegów__cat',   
          'param_moc','feature_kamera-cofania__cat',
          'param_typ__cat','seller_name__cat',   
          'feature_światła-led__cat',
          'feature_wspomaganie-kierownicy__cat',  
          'feature_system-start-stop__cat',
          'param_pojemność-skokowa',
          'feature_regulowane-zawieszenie__cat',
          'feature_łopatki-zmiany-biegów__cat',
          'feature_asystent-pasa-ruchu__cat',
          'feature_hud-(wyświetlacz-przezierny)__cat',
          'feature_czujniki-parkowania-przednie__cat',
          'param_marka-pojazdu__cat',
          'param_kod-silnika__cat' ] 
#feats =  ['param_napęd__cat','param_stan__cat','param_faktura-vat__cat','param_rok-produkcji','param_skrzynia-biegów__cat','param_moc','feature_kamera-cofania__cat','param_typ__cat','seller_name__cat','feature_światła-led__cat','feature_wspomaganie-kierownicy__cat','feature_system-start-stop__cat','param_pojemność-skokowa','feature_regulowane-zawieszenie__cat','feature_łopatki-zmiany-biegów__cat','feature_asystent-pasa-ruchu__cat','feature_hud-(wyświetlacz-przezierny)__cat','feature_czujniki-parkowania-przednie__cat','param_marka-pojazdu__cat','param_kod-silnika__cat' ] 

xgb_params = {
    'max_depth' : 5,
    'n_estimators' : 50,
    'learning_rate' : 0.1,
    'seed' : 0
}
run_model(xgb.XGBRegressor(**xgb_params), feats)
```

## Hyperopt
Poprawienie parametrów

```
def obj_func(params):
  print('Training with params: ')
  print(params)

  mean_mae, score_std = run_model(xgb.XGBRegressor(**params), feats)
  return {'loss':np.abs(mean_mae), 'status': STATUS_OK}

#space
xgb_reg_params = {
    'learning_rate':     hp.choice('learning_rate',     np.arange(0.05, 0.31, 0.05)),
    'max_depth':         hp.choice('max_depth',         np.arange(5, 17, 1, dtype=int)),
    'subsample':         hp.quniform('subsample', 0.5, 1, 0.05),
    'colsample_bytree':  hp.quniform('colsample_bytree', 0.55 , 1, 0.05),
    'objective':         'reg:squarederror', 
    'n_estimators':      100,
    'seed':              0,
}

#run
best = fmin(obj_func, xgb_reg_params, algo=tpe.suggest, max_evals=11)
```
